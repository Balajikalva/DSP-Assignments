# Convolution Of Two Sequences

## Function for possible pairs

```python
def solution(n,l1,l2):
    
    lst = list()
    
    for i in range(n+1):
        
        t = (i,n-i)
        
        if t[0] <= l1-1  and t[1] <= l2-1: 
            
            lst.append(t)
            
    return lst  
        
solution(7,8,9)
```

The function gives all the possible pairs for y[n] in convolution

## Sum and Product of complex numbers|

```python
## Given sequences are complex in nature so we need to write functions for finding sum and product of them

def multiply(a,b):
    
    x1,y1 = a
    
    x2,y2 = b
    
    ans0 = (x1*x2)-(y1*y2)
    
    ans1 = (x1*y2)+(y1*x2)
    
    ans = (ans0,ans1)
    
    return ans

c1 = (1,8)

c2 = (4,7)

print("product of {} and {} is :{}".format(c1,c2,multiply(c1,c2)))

def summer(a,b):
    
    x1,y1 = a
    
    x2,y2 = b
    
    return (x1+x2,y1+y2)

    
print("sum of {} and {} is :{}".format(c1,c2,summer(c1,c2)))
    
```

```python
s = 0

e = 10

import random

def my_conv(s1,s2):
    
    lst = list()
    
    l1 = len(s1)
    
    l2 = len(s2)
    
    n = max(l1,l2)
    
    for i in range(l1+l2-1):
        
        l = solution(i,l1,l2) # for finding all the possible combinations of i used for convolution
        
       # print(l)
        
        total = 0,0
        
        for items in l:
            
            v1 = items[0] # first value in this corresponds to sequence1 
            
            v2 = items[1] # second value corresponds to sequence2
            
            local = multiply(s1[v1],s2[v2]) # for multiplying values in s1 and s2
            
            total = summer(total,local) # summing of two complex numbers
            
        lst.append(total)
        
    return lst

#t = int(input("Enter the Length of sequence:\n"))

t = 8

s1_r = list()
    
s1_c = list()
    
s2_r = list()
    
s2_c = list()
    
for j in range(t):
        
    s1_r.append(random.randint(s,e))
        
    s1_c.append(random.randint(s,e))
        
    s2_r.append(random.randint(s,e))
        
    s2_c.append(random.randint(s,e))
        
s1 = list(zip(s1_r,s1_c))
    
s2 = list(zip(s2_r,s2_c))

print("\nFirst Sequence:\n",s1)

print("\nSecond Sequence:\n",s2)

ans = my_conv(s1,s2)

print("\nConvolution Result:\n",ans)

```

## Creating random data for given length of sequence

```python
l = list()

for i in range(1,31):
    
    l.append(i*10)
    
print(l)
```

```python
rough = list()

for j in range(10):
    
    rough.append(random.randint(s,e))

rough
```

```python
data = list()

for i in l:
    
    s1_r = list()
    
    s1_c = list()
    
    s2_r = list()
    
    s2_c = list()
    
    for j in range(i):
        
        s1_r.append(random.randint(s,e))
        
        s1_c.append(random.randint(s,e))
        
        s2_r.append(random.randint(s,e))
        
        s2_c.append(random.randint(s,e))
        
    s1 = list(zip(s1_r,s1_c))
    
    s2 = list(zip(s2_r,s2_c))
    
    data.append((s1,s2))
    

for i in range(2):
    
    print(data[i][0])
    
    print(data[i][1])
        
```

Data has the randomly generated data of sequences with lengths from 10 to 300 in the multiples of 10 


## Executing the convolution on s1,s2 10 times

```python
import time

for i in data:
    
    start_time = time.process_time()
    
    for k in range(10):
        
        my_conv(i[0],i[1])
        
        
    print(len(i[0]),time.process_time()-start_time)
    
```

## Executing the convolution on s1,s2 only 1time

```python
print('length    timetaken')

for i in data:
    
    start_time = time.process_time()
   
        
    my_conv(i[0],i[1])
        
        
    print(len(i[0]),'\t',time.process_time()-start_time)
```

For sequences of length from 10 to 80 , the execution time is very less , so we find the average over 20 in the next-step to be in the safer side and for correct estimation

```python
time_taken = list()

length = list()

for i in data:
    
    start_time = time.process_time()
    
    for k in range(50):
        
        my_conv(i[0],i[1])
    
    length.append(len(i[0]))
    
    t = (time.process_time()-start_time)/50
    
    time_taken.append(t)
    
```

```python
for i in range(30):
    
    print(length[i],time_taken[i])
```

length of the sequence and time-taken for exceution are stored in the lists length and time_taken
